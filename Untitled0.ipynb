```python
# ==============================
# ðŸ“¦ Sistem Rekomendasi Produk Berdasarkan Ulasan
# ==============================

# --- Instalasi pustaka yang dibutuhkan ---
!pip install pandas scikit-learn numpy

# --- Import library ---
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# --- Contoh dataset (bisa diganti dengan data kamu sendiri) ---
data = {
    'product_id': [1, 2, 3, 4, 5],
    'product_name': [
        'Laptop Lenovo',
        'Laptop HP',
        'Headphone Sony',
        'Mouse Logitech',
        'Keyboard Mechanical'
    ],
    'review': [
        'Laptop ini cepat dan tahan lama, sangat puas!',
        'Performa bagus tapi baterai cepat habis.',
        'Suara jernih dan bass mantap, kualitas tinggi.',
        'Nyaman dipakai, klik-nya lembut dan responsif.',
        'Tombol empuk dan cocok untuk gaming.'
    ]
}

df = pd.DataFrame(data)
print("ðŸ“‹ Data produk:")
display(df)

# --- TF-IDF Vectorization ---
tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(df['review'])

# --- Hitung kemiripan antar produk ---
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)

# --- Fungsi rekomendasi ---
def rekomendasi_produk(nama_produk, top_n=3):
    # cari index produk
    idx = df.index[df['product_name'] == nama_produk][0]
    # ambil skor kemiripan
    sim_scores = list(enumerate(cosine_sim[idx]))
    # urutkan berdasarkan skor (produk paling mirip di atas)
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    # ambil top_n produk mirip (selain dirinya sendiri)
    sim_scores = sim_scores[1:top_n+1]
    product_indices = [i[0] for i in sim_scores]
    return df[['product_name', 'review']].iloc[product_indices]

# --- Contoh penggunaan ---
produk_input = 'Laptop Lenovo'
print(f"ðŸŽ¯ Rekomendasi produk mirip dengan '{produk_input}':\n")
display(rekomendasi_produk(produk_input))
```
